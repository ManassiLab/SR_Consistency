**Other functions**

```python
#This function split a list in two halves
def split_list(a_list):
    half = len(a_list)//2
    return a_list[:half], a_list[half:]

#This function make a list of lists become a list of elements (e.g., [[a],[b]]-->[a,b])    
def flatten(seq): 
    l = []
    for elt in seq:
        t = type(elt)
        if t is tuple or t is list:
            for elt2 in flatten(elt):
                l.append(elt2)
        else:
            l.append(elt)
    return l
```

**Between-Consistency Analysis**

```python
### Between-subject consistency--> Bootstrap method to estimate split-half correlations 


#On each iteration, for each observer and each binned morph, randomly split the responses into two halves 

MEAN_Tot_halves_correlations = []

for b in range (1000):   
    Tot_MEANbin_1 = []
    Tot_MEANbin_2 = []
    NAN_ = 0
    for x in range(len(Subj)):
        mean_bin1_singleSubject = []
        mean_bin2_singleSubject = []
        for bin in range(49):
            #randomly shuffle the values in the bin
            actual_bin = (TotBins_allsubjects[x])[bin]
            random.shuffle(actual_bin)

            #randomly split the responses into two halves 
            a, b = split_list(actual_bin)

            #alert if a bin is = [] 
            
            if a == [] and b == []:
                a = 0
                b = 0
                #print("NANN")
                #print(bin)
           #if a bin = [], repeat the value of the other bin (this is to avoid NaN)
            if a == []:
                a = b
                NAN_ = NAN_ + 1
                #print("NAN")
                
            if b == []:
                b = a
                NAN_ = NAN_ + 1
                #print("NAN")
              

            #calculate the mean response errors for each half
            mean_bin1 = mean(a)
            mean_bin2 = mean(b)

            mean_bin1_singleSubject.append(mean_bin1)
            mean_bin2_singleSubject.append(mean_bin2)
            
        #Save in a list the data from all the subjects from bin 1 and 2   
        Tot_MEANbin_1.append(mean_bin1_singleSubject)
        Tot_MEANbin_2.append(mean_bin2_singleSubject)
    #print(len(Tot_MEANbin_1))
    
    #Correlate one half from one observer with one half from another--> Pearson’s r value
    #Shuffle lists of bin means
    random.shuffle(Tot_MEANbin_1)
    random.shuffle(Tot_MEANbin_2)
    
 
    Tot_halves_correlations = []
    for x in range(len(Tot_MEANbin_1)):

        #The two halves are correlated--> Pearson’s r value  
        corr, _ = pearsonr(Tot_MEANbin_1[x], Tot_MEANbin_2[x])
        
        #trasform Pearson's r values in Fisher z
        corr = np.arctanh(corr)
        
        #Append z values
        Tot_halves_correlations.append(corr) 
    #print(len(Tot_halves_correlations))
    
   #Trasform the average of Fisher z values in r Pearson's value
    Tot_halves_correlations_r = np.tanh(mean(Tot_halves_correlations))
    
    MEAN_Tot_halves_correlations.append(mean(Tot_halves_correlations_r))
    

print("NUMERO DI NAN: ", NAN_)    
  
print("\nAverage group Pearson's r value: ", mean(MEAN_Tot_halves_correlations))
```

**Plotting Data Between-Subject Consistency_average Between group** 

```python
#PLOT BAR GRAPH BETWEEN-SUBJECT CONSISTENCY MEAN

meanAmp = np.mean(MEAN_Tot_halves_correlations)
conf_interval1 = np.percentile(MEAN_Tot_halves_correlations,[2.5])
conf_interval2 = np.percentile(MEAN_Tot_halves_correlations,[97.5])


list1 = [1]

fig1 = plt.figure(figsize=(10,8))

plt.bar(list1, meanAmp, width=0.8, align='center')

plt.axis([0, 2, -0.1, 0.8]) 
#plt.xticks(["Control group"])
#plt.yticks(np.arange(0, 60, 5))

plt.errorbar(list1, meanAmp, yerr=(conf_interval1-meanAmp, meanAmp-conf_interval2), fmt=' ', ecolor='black', capsize=5)

title = "BETWEEN-SUBJECT CONSISTENCY\n" + "Mean: " + str(np.round(meanAmp, decimals = 2))  
plt.title(title, fontsize = 20)
plt.xlabel("\nRadiologist group", fontsize = 20)
plt.ylabel("Pearson's R", fontsize = 20)
plt.tick_params(labelsize = 15 ,bottom=False, labelbottom=False)
plt.legend()

plt.show()

FigName = CurrentPath + "/all_graphs/Bootstrap_Between-subjects_BarGraph_3bin.png" 

fig1.tight_layout()
fig1.savefig(FigName, dpi=600)

plt.clf()

```
